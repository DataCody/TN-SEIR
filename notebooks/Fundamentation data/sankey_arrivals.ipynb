```python
import pandas as pd
import random
import plotly.graph_objects as go
```

```python
df = pd.read_excel("arrive by region.xlsx")
df.head()
```

```python
print(df.columns)
```

```python
# Convert column names to string type
df.columns = df.columns.map(str)

# Prepare data for the Sankey diagram
nodes = list(df['Country'].unique()) + list(df['Arrivals from regions'].unique())
node_indices = {node: i for i, node in enumerate(nodes)}

# Define links (flows)
targets = [node_indices[country] for country in df['Country']]
sources = [node_indices[region] for region in df['Arrivals from regions']]
values = df['2019'].tolist()  # Use 2019 data as the flow values

# Generate random colors
def random_color():
    return f'rgba({random.randint(0, 255)}, {random.randint(0, 255)}, {random.randint(0, 255)}, 0.8)'

colors = [random_color() for _ in range(len(values))]

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=colors  # Set random colors for the links
    )
)])

# Set chart dimensions
fig.update_layout(
    #title_text="Sankey Plot: Arrivals by Region in 2019",
    font_size=16,
    width=800,  # Set the width of the chart
    height=600,  # Set the height of the chart
    paper_bgcolor='white',  # Set the background color of the paper
    plot_bgcolor='white',   # Set the background color of the plot area
    font=dict(color='black')  # Set the font color
)

fig.write_image('arrivals_regions.png', height=600, width=800, scale=3)

fig.show()
```

```python
import pandas as pd
import random
import plotly.graph_objects as go

df = pd.read_excel("arrive by region.xlsx")
df.columns = df.columns.map(str)

# Prepare data for the Sankey diagram
# Reorder the nodes to specify the desired order on the right side
nodes = list(df['Arrivals from regions'].unique()) + ['Fiji', 'French Polynesia', 'Samoa']  # Desired order

# Create a dictionary to map each node to an index
node_indices = {node: i for i, node in enumerate(nodes)}

# Define links (flows)
targets = [node_indices[country] for country in df['Country']]
sources = [node_indices[region] for region in df['Arrivals from regions']]
values = df['2019'].tolist()  # Use 2019 data as the flow values

# Generate random colors
def random_color():
    return f'rgba({random.randint(0, 255)}, {random.randint(0, 255)}, {random.randint(0, 255)}, 0.8)'

colors = [random_color() for _ in range(len(values))]

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
        color='lightgrey'  # Optional: color nodes to help visualize them better
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=colors
    )
)])

# Set chart dimensions
fig.update_layout(
    font_size=16,
    width=800,  # Set the width of the chart
    height=600,  # Set the height of the chart
    paper_bgcolor='white',  # Set the background color of the paper
    plot_bgcolor='white',   # Set the background color of the plot area
    font=dict(color='black')  # Set the font color
)

fig.write_image('arrivals_regions.png', height=600, width=800, scale=3)

fig.show()
```
