```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the provided Excel file
file_path = 'arrive by region.xlsx'
data = pd.read_excel(file_path)

# Convert column names to strings
data.columns = data.columns.astype(str)

# Filter data for each country and region
countries = data['Country'].unique()
regions = data['Arrivals from regions'].unique()

# Create a dataframe to hold the sums for each country and region
region_data = pd.pivot_table(data, values=['2022'], index=['Country', 'Arrivals from regions'], aggfunc='sum')

# Prepare data for the bar chart
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.2
x = np.arange(len(regions))

# Plot bars for each country
for i, country in enumerate(countries):
    region_sums = region_data.loc[country].sum(axis=1).values
    ax.bar(x + i * bar_width, region_sums, bar_width, label=country)

# Add labels, title, and legend
ax.set_xlabel('Region')
ax.set_ylabel('Arrivals (Thousands)')
ax.set_title('Arrivals by Region in 2022')
ax.set_xticks(x + bar_width)
ax.set_xticklabels(regions, rotation=45, ha='right')
ax.legend()

fig.tight_layout()
plt.savefig('arrivals by regions.png', bbox_inches='tight', dpi=300)
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the provided Excel file
file_path = 'arrive by region.xlsx'
data = pd.read_excel(file_path)

# Convert column names to strings
data.columns = data.columns.astype(str)

# Filter data for each country and region
countries = data['Country'].unique()
regions = data['Arrivals from regions'].unique()

# Create a dataframe to hold the sums for each country and region
region_data = pd.pivot_table(data, values=['2022'], index=['Country', 'Arrivals from regions'], aggfunc='sum')

# Calculate total arrivals for each country and sort
country_totals = region_data.groupby('Country').sum().sum(axis=1)
sorted_countries = country_totals.sort_values(ascending=False).index

# Prepare data for the bar chart, sorted by total arrivals
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.2
x = np.arange(len(regions))

# Plot bars for each country in sorted order
for i, country in enumerate(sorted_countries):
    region_sums = region_data.loc[country].sum(axis=1).values
    ax.bar(x + i * bar_width, region_sums, bar_width, label=country)

# Add labels, title, and legend
ax.set_xlabel('Region')
ax.set_ylabel('Arrivals (Thousands)')
ax.set_title('Arrivals by Region in 2022 (Sorted by Total)')
ax.set_xticks(x + bar_width * (len(sorted_countries) - 1) / 2)  # Adjust xticks position for sorted bars
ax.set_xticklabels(regions, rotation=45, ha='right')
ax.legend()

fig.tight_layout()
plt.savefig('arrivals_by_regions_sorted.png', bbox_inches='tight', dpi=300)
# Display the plot
plt.show()
```

```python
import plotly.express as px
import pandas as pd

#  Excel 
file_path = 'hotmap.xlsx'
df = pd.read_excel(file_path)

# 
df = df[['Country', 'month', 'Total']]

# 
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# 
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)

# 
pivot_table = df.pivot(index='Country', columns='month', values='Total')

#  Plotly Express 
fig = px.imshow(pivot_table,
                color_continuous_scale='YlGnBu',
                labels={'x': 'Month', 'y': 'Country', 'color': 'Total'},
                text_auto='.0f')#title='Country Total by Month',

# 
fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Country',
    xaxis=dict(
        tickmode='array',
        tickvals=pivot_table.columns,
        ticktext=pivot_table.columns,
        title_font_size=18,
        tickfont_size=16
    ),
    yaxis=dict(
        tickmode='array',
        tickvals=pivot_table.index,
        ticktext=pivot_table.index,
        title_font_size=18,
        tickfont_size=16
    ),
    coloraxis_colorbar=dict(
        title='Total',
        title_font_size=18,
        tickfont_size=16,
        orientation='h',  # 
        xanchor='center',  # 
        yanchor='top',  # 
        x=0.5,  # 
        y=0  # 
    ),
    width=1200,  # Width of the figure
    height=600,  # Height of the figure
    margin=dict(l=10, r=30, t=10, b=5)  # Adjust margins to make space for the color bar
)
fig.write_image('vistors_hotmap.png', height=600, width=1200, scale=3)
# 
fig.show()
```
