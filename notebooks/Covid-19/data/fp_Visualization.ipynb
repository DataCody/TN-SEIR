```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', 'D', 'd', '+', 'x', '|', '_']

plt.figure(figsize=(10, 6))

for i, marker in enumerate(markers):
    plt.plot(x, [value + i for value in y], label=f'marker {marker}', marker=marker)

plt.legend()
plt.show()
```

### Simple SEIR vs. Travel Network SEIR

```python
simple_seir = pd.read_csv('fp/fp_simple_seir.csv')
no_measure = pd.read_csv('fp/fp_no_measure.csv')
```

```python
simple_seir['Day'] = simple_seir['Day'] + 1
print(simple_seir.head())
```

```python
pd.set_option('display.expand_frame_repr', False)
print(no_measure.head())
```

```python
# Plot S
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(simple_seir['Day'], simple_seir['S'] / 1000, label='Traditional SEIR', marker='o', markevery=30)
plt.plot(no_measure['Day'], no_measure['S'] / 1000, label='Travel Network SEIR', marker='x', markevery=30)
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Susceptible')
plt.legend(fontsize=6)

# Plot E
plt.subplot(2, 2, 2)
plt.plot(simple_seir['Day'], simple_seir['E'] / 1000, label='Traditional SEIR', marker='o', markevery=30)
plt.plot(no_measure['Day'], no_measure['E'] / 1000, label='Travel Network SEIR', marker='x', markevery=30)
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Exposed')
plt.legend(fontsize=6)

# Plot I
plt.subplot(2, 2, 3)
plt.plot(simple_seir['Day'], simple_seir['I'] / 1000, label='Traditional SEIR', marker='o', markevery=30)
plt.plot(no_measure['Day'], no_measure['I'] / 1000, label='Travel Network SEIR', marker='x', markevery=30)
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Infectious')
plt.legend(fontsize=6)

# Plot R
plt.subplot(2, 2, 4)
plt.plot(simple_seir['Day'], simple_seir['R'] / 1000, label='Traditional SEIR', marker='o', markevery=30)
plt.plot(no_measure['Day'], no_measure['R'] / 1000, label='Travel Network SEIR', marker='x', markevery=30)
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Recovered')
plt.legend(fontsize=6)

plt.tight_layout()
plt.show()
```

### Lockdown

```python
lockdown_outsideworld = pd.read_csv('fp/fp_lockdown_outsideworld.csv')
lockdown_outside_tahiti = pd.read_csv('fp/fp_lockdown_outside_tahiti.csv')
lockdownd_tahiti_borabora = pd.read_csv('fp/fp_lockdownd_tahiti_borabora.csv')
lockdown_all = pd.read_csv('fp/fp_lockdown_all.csv')
```

```python
# Plot S
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(lockdown_outsideworld['Day'], lockdown_outsideworld['S'] / 1000, label='Outside World', marker='o', markevery=30)
plt.plot(lockdown_outside_tahiti['Day'], lockdown_outside_tahiti['S'] / 1000, label='Largest Location', marker='x', markevery=30)
plt.plot(lockdownd_tahiti_borabora['Day'], lockdownd_tahiti_borabora['S'] / 1000, label='Popular Locations', marker='d', markevery=30)
plt.plot(lockdown_all['Day'], lockdown_all['S'] / 1000, label='All Locations', marker='s', markevery=30)
# Highlight the intervention period
method_start = 10
method_end = 100
plt.axvspan(method_start, method_end, color='grey', alpha=0.5, label='Lockdown Period')
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Susceptible')
plt.legend(fontsize=6)

# Plot E
plt.subplot(2, 2, 2)
plt.plot(lockdown_outsideworld['Day'], lockdown_outsideworld['E'] / 1000, label='Outside World', marker='o', markevery=30)
plt.plot(lockdown_outside_tahiti['Day'], lockdown_outside_tahiti['E'] / 1000, label='Largest Location', marker='x', markevery=30)
plt.plot(lockdownd_tahiti_borabora['Day'], lockdownd_tahiti_borabora['E'] / 1000, label='Popular Locations', marker='d', markevery=30)
plt.plot(lockdown_all['Day'], lockdown_all['E'] / 1000, label='All Locations', marker='s', markevery=30)
# Highlight the intervention period
method_start = 10
method_end = 100
plt.axvspan(method_start, method_end, color='grey', alpha=0.5, label='Lockdown Period')
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Exposed')
plt.legend(fontsize=6)

# Plot I
plt.subplot(2, 2, 3)
plt.plot(lockdown_outsideworld['Day'], lockdown_outsideworld['I'], label='Outside World', marker='o', markevery=30)
plt.plot(lockdown_outside_tahiti['Day'], lockdown_outside_tahiti['I'], label='Largest Location', marker='x', markevery=30)
plt.plot(lockdownd_tahiti_borabora['Day'], lockdownd_tahiti_borabora['I'], label='Popular Locations', marker='d', markevery=30)
plt.plot(lockdown_all['Day'], lockdown_all['I'], label='All Locations', marker='s', markevery=30)
# Highlight the intervention period
method_start = 10
method_end = 100
plt.axvspan(method_start, method_end, color='grey', alpha=0.5, label='Lockdown Period')
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Infectious')
plt.legend(fontsize=6)

# Plot R
plt.subplot(2, 2, 4)
plt.plot(lockdown_outsideworld['Day'], lockdown_outsideworld['R'], label='Outside World', marker='o', markevery=30)
plt.plot(lockdown_outside_tahiti['Day'], lockdown_outside_tahiti['R'], label='Largest Location', marker='x', markevery=30)
plt.plot(lockdownd_tahiti_borabora['Day'], lockdownd_tahiti_borabora['R'], label='Popular Locations', marker='d', markevery=30)
plt.plot(lockdown_all['Day'], lockdown_all['R'], label='All Locations', marker='s', markevery=30)
# Highlight the intervention period
method_start = 10
method_end = 100
plt.axvspan(method_start, method_end, color='grey', alpha=0.5, label='Lockdown Period')
plt.xlabel('Day')
plt.ylabel('Population (thousand)')
plt.title('Recovered')
plt.legend(fontsize=6)

plt.tight_layout()
plt.show()
```
