```python
import networkx as nx
import matplotlib.pyplot as plt
import json

plt.figure(figsize=(18, 12))

json_path = 'fj_network.json'
with open(json_path, 'r') as file:
    df = json.load(file)

locations = df['n']
links = df['Connectivity Matrix W']
ports = df['ports_according_communities']

location_map = {}
for location, port_list in ports.items():
    for port in port_list:
        location_map[port.strip()] = location


G = nx.Graph()

for location, population in locations.items():
    G.add_node(location, size=population, label=f"{population}")

# volume
for connection, volume in links.items():
    loc1, loc2 = [location_map[port.strip()] for port in connection.split(',')]
    G.add_edge(loc1, loc2, weight=volume, label=f"{volume}")


pos = nx.spring_layout(G)
sizes = [G.nodes[node]['size'] / 300 for node in G]
weights = [G[u][v]['weight'] / 300 for u, v in G.edges()]

nx.draw_networkx_nodes(G, pos, node_size=sizes, alpha=0.6)
nx.draw_networkx_edges(G, pos, width=weights, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8)

for node, (x, y) in pos.items():
    plt.text(x, y, s=G.nodes[node]['label'], horizontalalignment='right', verticalalignment='bottom', fontweight='bold')


for (u, v, attr) in G.edges(data=True):
    weight = attr['weight']
    mid_x, mid_y = (pos[u][0] + pos[v][0]) / 2, (pos[u][1] + pos[v][1]) / 2
    plt.text(mid_x, mid_y, s=f"{weight}", fontweight='bold', color='black')

plt.title('Graph of Fiji')
plt.show()
```

```python
import plotly.graph_objects as go
import networkx as nx

def draw_plotly_graph(json_path, title):
    with open(json_path, 'r') as file:
        df = json.load(file)

    locations = df['n']
    links = df['Connectivity Matrix W']
    ports = df['ports_according_communities']

    location_map = {}
    for location, port_list in ports.items():
        for port in port_list:
            location_map[port.strip()] = location

    G = nx.Graph()

    for location, population in locations.items():
        G.add_node(location, size=population, label=f"{location} ({population})")

    for connection, volume in links.items():
        loc1, loc2 = [location_map[port.strip()] for port in connection.split(',')]
        G.add_edge(loc1, loc2, weight=volume)

    pos = nx.spring_layout(G)
    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])

    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=0.5, color='#888'),
        hoverinfo='none',
        mode='lines')

    node_x = []
    node_y = []
    node_text = []
    node_size = []
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        node_text.append(G.nodes[node]['label'])
        node_size.append(G.nodes[node]['size'])

    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        hoverinfo='text',
        text=node_text,
        marker=dict(
            showscale=True,
            colorscale='YlGnBu',
            reversescale=True,
            color=node_size,
            size=[size / 1000 for size in node_size],
            colorbar=dict(
                thickness=15,
                title='Node Size',
                xanchor='left',
                titleside='right'
            ),
            line_width=2))

    fig = go.Figure(data=[edge_trace, node_trace],
                    layout=go.Layout(
                        title=title,
                        titlefont_size=16,
                        showlegend=False,
                        hovermode='closest',
                        margin=dict(b=20, l=5, r=5, t=40),
                        annotations=[dict(
                            text="Python code: <a href='https://plotly.com/python/network-graphs/'> https://plotly.com/python/network-graphs/</a>",
                            showarrow=False,
                            xref="paper", yref="paper",
                            x=0.005, y=-0.002)],
                        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                    )
    fig.show()

# Draw graphs for each network
draw_plotly_graph('fj_network.json', 'Graph of Fiji')
draw_plotly_graph('fp_network.json', 'Graph of French Polynesia')
draw_plotly_graph('ws_network.json', 'Graph of Samoa')
```
