```python
import networkx as nx
import json
import os

def calculate_community_strength(json_path):
    if not os.path.isfile(json_path):
        raise FileNotFoundError(f"File not found: {json_path}")
    
    with open(json_path, 'r', encoding='utf-8') as file:
        df = json.load(file)

    locations = df['n']
    links = df['Connectivity Matrix W']
    ports = df['ports_according_communities']

    location_map = {}
    for location, port_list in ports.items():
        for port in port_list:
            location_map[port.strip()] = location

    G = nx.Graph()

    for location, population in locations.items():
        G.add_node(location, size=population, label=f"{location}")

    for connection, volume in links.items():
        loc1, loc2 = [location_map[port.strip()] for port in connection.split(',')]
        G.add_edge(loc1, loc2, weight=volume)

    # Add community attribute to each node
    for community, ports_list in ports.items():
        for port in ports_list:
            node = location_map.get(port.strip())
            if node:
                G.nodes[node]['community'] = community

    community_strength = {}
    
    for node in G.nodes():
        community = G.nodes[node].get('community', 'Unknown')  # Get community info
        if community not in community_strength:
            community_strength[community] = 0
        
        for neighbor in G.neighbors(node):
            if G.has_edge(node, neighbor):
                edge_weight = G[node][neighbor]['weight']
                if node < neighbor:  # Avoid double counting
                    community_strength[community] += edge_weight
    
    return community_strength

# Calculate and print total connection strength for each community
def print_community_strength(json_file):
    community_strength = calculate_community_strength(json_file)
    for community, strength in community_strength.items():
        print(f"Community: {community}, Total Strength: {strength}")

# Print strengths for each network file
print("Fiji Network:")
print_community_strength('fj_network.json')

print("\nFrench Polynesia Network:")
print_community_strength('fp_network.json')

print("\nSamoa Network:")
print_community_strength('ws_network.json')
```
