```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## SEIR

```python
fj_seir = pd.read_csv('data/fj/fj_simple_seir.csv')
fp_seir = pd.read_csv('data/fp/fp_simple_seir.csv')
ws_seir = pd.read_csv('data/ws/ws_simple_seir.csv')
fj_seir['Day'] = fj_seir['Day'] + 1
fp_seir['Day'] = fp_seir['Day'] + 1
ws_seir['Day'] = ws_seir['Day'] + 1
```

## TN-SEIR

```python
fj_tNet = pd.read_csv('data/fj/fj_no_measure.csv')
fp_tNet = pd.read_csv('data/fp/fp_no_measure.csv')
ws_tNet = pd.read_csv('data/ws/ws_no_measure.csv')
```

## SEIR vs. TN-SEIR

```python
plt.figure(figsize=(12, 4))

# Plot Fiji
plt.subplot(1, 3, 1)
plt.plot(fj_seir['Day'] / 7, fj_seir['I'] / 1000, marker='o', markevery=30, label='SEIR')
plt.plot(fj_tNet['Day'] / 7, fj_tNet['I'] / 1000, marker='x', markevery=30, label='TN-SEIR')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Fiji', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot French Polynesia
plt.subplot(1, 3, 2)
plt.plot(fp_seir['Day'] / 7, fp_seir['I'] / 1000, marker='o', markevery=30, label='SEIR')
plt.plot(fp_tNet['Day'] / 7, fp_tNet['I'] / 1000, marker='x', markevery=30, label='TN-SEIR')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('French Polynesia', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot Samoa
plt.subplot(1, 3, 3)
plt.plot(ws_seir['Day'] / 7, ws_seir['I'] / 1000, marker='o', markevery=30, label='SEIR')
plt.plot(ws_tNet['Day'] / 7, ws_tNet['I'] / 1000, marker='x', markevery=30, label='TN-SEIR')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Samoa', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

lines, labels = plt.subplot(1, 3, 1).get_legend_handles_labels()
plt.figlegend(lines, labels, loc='lower center', ncol=3, fontsize=10)
# Adjust layout and save figure
plt.tight_layout(rect=[0, 0.05, 1, 0.96])
plt.savefig('three_countries_seir_tn-seir.png', dpi=300)
plt.show()
```

## Lockdown

```python
# No Measure
fj_no_measure = pd.read_csv('data/fj/fj_no_measure.csv')
fp_no_measure = pd.read_csv('data/fp/fp_no_measure.csv')
ws_no_measure = pd.read_csv('data/ws/ws_no_measure.csv')
```

```python
# Fiji Lockdown
fj_outside = pd.read_csv('data/fj/fj_lockdown_outsideworld.csv')
fj_city = pd.read_csv('data/fj/fj_lockdown_outside_suva.csv')
fj_connection = pd.read_csv('data/fj/fj_lockdownd_suva_nadi.csv')
fj_all = pd.read_csv('data/fj/fj_lockdown_all.csv')
```

```python
# French Polynesia Lockdown
fp_outside = pd.read_csv('data/fp/fp_lockdown_outsideworld.csv')
fp_city = pd.read_csv('data/fp/fp_lockdown_outside_tahiti.csv')
fp_connection = pd.read_csv('data/fp/fp_lockdownd_tahiti_borabora.csv')
fp_all = pd.read_csv('data/fp/fp_lockdown_all.csv')
```

```python
# Samoa Lockdown
ws_outside = pd.read_csv('data/ws/ws_lockdown_outsideworld.csv')
ws_city = pd.read_csv('data/ws/ws_lockdown_outside_falealili.csv')
ws_connection = pd.read_csv('data/ws/ws_lockdownd_Faleata_Vaimauga_Sagaga.csv')
ws_all = pd.read_csv('data/ws/ws_lockdown_all.csv')
```

```python
plt.figure(figsize=(18, 20))

# 
no_intervention_color = 'b'  # 
intervention_color = 'r'  # 
no_intervention_marker = 'x'
intervention_marker = 'o'
shade_color = 'grey'
shade_alpha = 0.5
method_start = 5
method_end = 19

# Lockdown Outside World
# Plot Fiji
plt.subplot(4, 3, 1)
plt.plot(fj_no_measure['Day'] / 7, fj_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fj_outside['Day'] / 7, fj_outside['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Fiji', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)
plt.text(-0.2, 0.5, 'Outside World', transform=plt.gca().transAxes, fontsize=16, fontweight='bold', va='center', rotation=90)

# Plot French Polynesia
plt.subplot(4, 3, 2)
plt.plot(fp_no_measure['Day'] / 7, fp_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fp_outside['Day'] / 7, fp_outside['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('French Polynesia', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot Samoa
plt.subplot(4, 3, 3)
plt.plot(ws_no_measure['Day'] / 7, ws_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(ws_outside['Day'] / 7, ws_outside['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Samoa', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Lockdown Main City
# Plot Fiji
plt.subplot(4, 3, 4)
plt.plot(fj_no_measure['Day'] / 7, fj_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fj_city['Day'] / 7, fj_city['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Fiji', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)
plt.text(-0.2, 0.5, 'Main City', transform=plt.gca().transAxes, fontsize=16, fontweight='bold', va='center', rotation=90)

# Plot French Polynesia
plt.subplot(4, 3, 5)
plt.plot(fp_no_measure['Day'] / 7, fp_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fp_city['Day'] / 7, fp_city['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('French Polynesia', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot Samoa
plt.subplot(4, 3, 6)
plt.plot(ws_no_measure['Day'] / 7, ws_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(ws_city['Day'] / 7, ws_city['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Samoa', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Lockdown Most Connections
# Plot Fiji
plt.subplot(4, 3, 7)
plt.plot(fj_no_measure['Day'] / 7, fj_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fj_connection['Day'] / 7, fj_connection['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Fiji', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)
plt.text(-0.2, 0.5, 'Most Connections', transform=plt.gca().transAxes, fontsize=16, fontweight='bold', va='center', rotation=90)

# Plot French Polynesia
plt.subplot(4, 3, 8)
plt.plot(fp_no_measure['Day'] / 7, fp_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fp_connection['Day'] / 7, fp_connection['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('French Polynesia', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot Samoa
plt.subplot(4, 3, 9)
plt.plot(ws_no_measure['Day'] / 7, ws_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(ws_connection['Day'] / 7, ws_connection['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Samoa', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Lockdown All
# Plot Fiji
plt.subplot(4, 3, 10)
plt.plot(fj_no_measure['Day'] / 7, fj_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fj_all['Day'] / 7, fj_all['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Fiji', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)
plt.text(-0.2, 0.5, 'All Locations', transform=plt.gca().transAxes, fontsize=16, fontweight='bold', va='center', rotation=90)

# Plot French Polynesia
plt.subplot(4, 3, 11)
plt.plot(fp_no_measure['Day'] / 7, fp_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(fp_all['Day'] / 7, fp_all['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('French Polynesia', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# Plot Samoa
plt.subplot(4, 3, 12)
plt.plot(ws_no_measure['Day'] / 7, ws_no_measure['I'] / 1000, marker=no_intervention_marker, markevery=30, label='No Intervention', color=no_intervention_color)
plt.plot(ws_all['Day'] / 7, ws_all['I'] / 1000, marker=intervention_marker, markevery=30, label='Intervention', color=intervention_color)
plt.axvspan(method_start, method_end, color=shade_color, alpha=shade_alpha, label='Intervention Period')
plt.xlabel('EpiWeek')
plt.ylabel('Infectious (thousand)')
plt.title('Samoa', fontsize=14, fontweight='bold')
plt.legend(fontsize=10).set_visible(False)

# 
lines, labels = plt.subplot(4, 3, 1).get_legend_handles_labels()
plt.figlegend(lines, labels, loc='lower center', ncol=3, fontsize=10)

plt.tight_layout(rect=[0, 0.01, 1, 0.96])
plt.savefig('locckdown_intervention.png', bbox_inches='tight', dpi=400)
plt.show()
```
